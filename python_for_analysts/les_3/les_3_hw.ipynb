```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasetes/kc_house_data.csv')
df.head()
```

```python
# 2. Добавить новый признак price_per_sqft_living, который будет содержать среднюю стоимость за квадратный метр жилой площади (price / sqft_living).
```

```python
df['price_per_sqft_living'] = df['price'] / df['sqft_living']
```

```python
df.head()
```

```python
# . Создать новый признак age_of_house, который будет содержать возраст дома (текущий год минус год постройки yr_built).
```

```python
df['age_of_house'] = 2024 - df['yr_built']
```

```python
df.head()
```

```python
df.head()
```

```python
# 1.4 Удалите признаки zipcode, lat, long
```

```python
df.drop(['zipcode', 'lat', 'long'], axis=1, inplace=True)
```

```python
# Создание новых признаков: год и месяц продажи
```

```python
df['year_ch'] = pd.to_datetime(df['date'],
format='%Y%m%dT000000').dt.year
```

```python
df['month_ch'] = pd.to_datetime(df['date'],
format='%Y%m%dT000000').dt.month
```
