### Построим матрицу корреляции для выявления линейной зависимоти

```python
# подготовим признаки
df_num_features = df.select_dtypes(include=['float64', 'int64'])
df_num_features.drop(['id','zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15'], axis=1, inplace=True)
```

```python
df_num_features.info()
```

```python
corr_matrix = df_num_features.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
```

```python
plt.figure(figsize=(10, 5))

sns.heatmap(corr_matrix, annot=True, linewidths=0.7, cmap='coolwarm')
plt.xticks(rotation=35)
# plt.yticks(rotation=35)
plt.show()
```

#### Рассмотрим матрицу и определим какие характеристики недвижимости влияют на стоимость недвижимости.

1. 0.7 - price / sqft_living
1. 0.7 - price / grade
1. 0.6 - price / sqft_above
1. 0.5 - price / bathrooms
1. 0.4 - price / view

#### Построим данные графики:

```python
price_in_ml = df['price'] / 1_000_000

plt.scatter(data=df, x='lat', y='long', c=price_in_ml, alpha=0.5, linewidths=1.5, s=80)
plt.xlabel('Широта')
plt.ylabel('Долгота')

plt.show()
print('Cтоимость жилья в ценрте города значительно выше')
```

```python
plt.scatter(data=df, x=price_in_ml, y='sqft_living', c=price_in_ml, alpha=0.7, linewidths=1.5, s=80)

plt.xlabel('Стоимость в мл.$')
plt.ylabel('Жилая площадь')
 
plt.show()
print('Чем больше площадь жилья, соответсвенно тем выше стоимость жилья')
```

```python
plt.scatter(data=df, x=price_in_ml, y='grade', c=price_in_ml, alpha=0.7, linewidths=1.5, s=80)

plt.xlabel('Стоимость в мл.$')
plt.ylabel('Качество конструкции и дизайна')

plt.show()
print('Чем выше качество конструкции и дизайна, тем выше стоимость жилья')
```

```python
plt.scatter(data=df, x=price_in_ml, y='view', c=price_in_ml,  alpha=0.7, linewidths=1.5, s=80)

plt.xlabel('Стоимость в мл.$')
plt.ylabel('Насколько хороший вид')

plt.show()
print('Наилучший вид из окон, влияет на стоимость жилья')
```

```python
plt.scatter(data=df, x=price_in_ml, c=price_in_ml, y='bathrooms', alpha=0.7, linewidths=1.5, s=80)

plt.xlabel('Стоимость в мл.$')
plt.ylabel('Количество ванных комнат')

plt.show()
print('Наличие нескольких ванных комнат, также влияет на стоимость жилья')
```

```python
plt.scatter(data=df, x=price_in_ml, c=price_in_ml, y='sqft_above', alpha=0.7, linewidths=1.5, s=80)

plt.xlabel('Стоимость в мл.$')
plt.ylabel('Кв. метры дома, выше земли')

plt.show()
print('Чем больше кв. мет дома, выше земли, пололижительно влияет на стоимость жилья')
```
