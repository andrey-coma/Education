```python
df.drop('ID', axis=1, inplace=True)
```

```python
df = df.reset_index()
```

```python
df.drop('index', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Club'].unique()
```

```python
Age_Group = ['up to 20', 'from 20 to 30', 'from 30 to 36', 'over 36']
```

```python
def age_group(value: int) -> str:
    if value < 20:
        return Age_Group[0]
    
    elif (value > 20) & (value < 30):
        return Age_Group[1]
    
    elif (value > 30) & (value < 36):
        return Age_Group[2]
    
    else:
        return Age_Group[3]
```

```python
df['Age Group'] = df['Age'].apply(age_group)
```

```python
df.head(5)
```

```python
pd.pivot_table(df, values='Name', columns='Age Group', aggfunc='count')
```
